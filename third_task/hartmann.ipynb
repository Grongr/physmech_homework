```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
eps0 = 8.8541878128e-12
mu0 = 1.25663706212e-6

H = 20
l = 1
sigma = 1
eta = 1
v0 = 1
dp = - v0 * eta / (l * l)

Ha = H * l * np.sqrt(sigma / eta)
shHa = np.sinh(Ha)
chHa = np.cosh(Ha)

K = np.sqrt(eta * sigma) * mu0
A = l * l * dp / (eta * Ha * shHa)
```

```python
z = np.linspace(-l, l, 100)
sh = np.sinh(Ha * z / l)
ch = np.cosh(Ha * z / l)

v = v0 * (sh + shHa) / (2 * shHa) + A * (ch - chHa)
j = -np.sqrt(eta * sigma) * Ha / l * ( v0 / (2 * shHa) * sh + A * (ch - shHa / Ha))
h = mu0 * np.sqrt(sigma * eta) * (v0 / (2 * shHa) * (ch - chHa) + A * (sh - z / l * shHa))

v /= v0
j /= np.sqrt(eta * sigma) * v0 / l
h /= mu0 * np.sqrt(sigma * eta) * v0
z = z / l

np.savetxt(str(H) + '.dat', np.array([z, v, j, h]).T)
```

```python

v = v0 / 2 * (1 + z / l) + l * l / 2 / eta * dp * (z * z / l / l - 1)
v /= v0
np.savetxt('0.dat', np.array([z, v]).T)
```
